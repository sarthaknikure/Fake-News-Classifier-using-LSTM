## Fake News Classifier using Bidirectional LSTM (NLP)

Problem Statement: We are going to use "Title" to classify our News is Fake or Real,So "title" is our Independent variable,whereas our target is to detect our News is Fake or Not,So here our dependent/target variable is label,So Let's get the Independent & Dependent variable

```python
## Drop nan values
df=df.dropna()
```

```python
## Get Independent Features
X=df.drop('label', axis=1)
```

```python
from typing import Sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
# Vocabulary Size
voc_size=5000
```

Onehot Representation

```python
messages=X.copy()
```

```python
messages[['title','author','text']]
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
## msg=messages.loc[:,['title','author','text']] #You can use loc to select all rows but only a specific number of columns
```

```python
## msg.iloc[1]   #The iloc function allows you to subset pandas DataFrames based on their position or index.
```

```python
## Dataset Preprocessing
from nltk.stem.porter import PorterStemmer #Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language.
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words, voc_size) for words in corpus]
onehot_repr
```

```python
sent_length=30
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
from tensorflow.keras.layers import Bidirectional
#from tensorflow.python.keras.layers.wrappers import Bidirectional
## Creating Model
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100)))
#model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
len(embedded_docs), y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
# Finally Training
model1.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=10, batch_size=64)
```

Performance Metrics and Accuracy

```python
y_pred1=(model1.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
